```python
video_schema = {
    "properties": {
        "video_description": {"type": "string"},
        "video_topic": {"type": "string"},
        "video_length": {"type": "string"},
        "steps_of_procedure": {"type": "string"},
        "start_time_of_step": {"type": "string"},
        "end_time_of_step": {"type": "string"}
    }
}
```

```python
'''
Step Schema for Bike Chain Repair
'''
step_schema = {
    "properties": {
        "Start Time: remove the old chain": {"type": "string"},
        "End Time: remove the old chain": {"type": "string"},
        "Transcript context: remove the old chain": {"type": "string"},

        "Start Time: clean the cassette": {"type": "string"},
        "End Time: clean the cassette": {"type": "string"},
        "Transcript context: clean the cassette": {"type": "string"},

        "Start Time: thread the new chain": {"type": "string"},
        "End Time: thread the new chain": {"type": "string"},
        "Transcript context: thread the new chain": {"type": "string"},

        "Start Time: determine the correct length of the new chain": {"type": "string"},
        "End Time: determine the correct length of the new chain": {"type": "string"},
        "Transcript context: determine the correct length of the new chain": {"type": "string"},

        "Start Time: cut the new chain to the desired length": {"type": "string"},
        "End Time: cut the new chain to the desired length": {"type": "string"},
        "Transcript context: cut the new chain to the desired length": {"type": "string"},

        "Start Time: insert the pin into the new chain": {"type": "string"},
        "End Time: insert the pin into the new chain": {"type": "string"},
        "Transcript context: insert the pin into the new chain": {"type": "string"},

        "Start Time: push the pin in": {"type": "string"},
        "End Time: push the pin in": {"type": "string"},
        "Transcript context: push the pin in": {"type": "string"},
    }
}
```

```python
"""## Schema extractions

### General video description schema extraction
"""
"""# Extraction chains for each type of schema extraction"""

# hopefully I can take this out soon
# from langchain.chains import create_extraction_chain
# chain_video = create_extraction_chain(video_schema, llm)

vid_KO = chain_video.run(inp)
# vid_KO

"""### Step schema extraction with tech pub steps"""

# chain_step = create_extraction_chain(step_schema, llm)

steps_KO = chain_step.run(inp)
# steps_KO

# Split steps_KO into dictionaries of each step

# New data structure for multiple dicts
new_steps_KO = []

# Only considers dictionary splitting if steps_KO output is one entire dictionary
if len(steps_KO) == 1:
    count = 0
    new_dict = {}
    for di in steps_KO:
        for key,val in di.items():
            new_dict[key] = val
            count += 1

            if count == 2:
                new_steps_KO.append(new_dict)
                new_dict = {}
                count = 0

# steps_KO is already split into dictionaries by tech pub step
else:
    new_steps_KO = steps_KO

new_steps_KO

"""## Output file on schema extraction results"""

with open('/Users/shrutisrinivasan/Desktop/HCI Code/KO_extraction_results/video_1.txt', mode='w') as f1:
    f1.write('Video 1 KO Extraction Results\n\n')
    f1.write(str(vid_KO))
    f1.write('\n\n')
    f1.write('Schema Extractions:\n')

    for di in new_steps_KO:
        for key,val in di.items():
            f1.write(key + ': ' + val)
            f1.write('\n')
```
